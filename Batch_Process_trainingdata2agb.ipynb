```python
# Fetch all csv files from gcs, use Xarray and dask, use gradient boosting ML, predict agb, 
# visualise it 
import os
import pandas as pd
import xarray as xr
import dask
import dask.dataframe as dd
from google.cloud import storage
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
import joblib
import json
from dask.distributed import Client
from sklearn.metrics import mean_squared_error

from dask.distributed import Client, LocalCluster

# Set up a local cluster with 4 workers
cluster = LocalCluster(n_workers=4)
client = Client(cluster)
print(client)

# Configure Google Cloud Storage (GCS) client
GCS_BUCKET_NAME = "test-agb-bucket"
GCS_FOLDER = "/agb_training_data"
LOCAL_DATA_DIR = "./data"
os.makedirs(LOCAL_DATA_DIR, exist_ok=True)

# Fetch CSV files from GCS
def download_csv_files():
    client = storage.Client()
    bucket = client.bucket(GCS_BUCKET_NAME)
    blobs = bucket.list_blobs()

    csv_files = []
    for blob in blobs:
        if blob.name.endswith(".csv") and not blob.name.endswith("155.csv"):
            local_path = os.path.join(LOCAL_DATA_DIR, os.path.basename(blob.name))
            blob.download_to_filename(local_path)
            csv_files.append(local_path)
    
    return csv_files

# Load data with Dask
def load_data(csv_files):
    dask_dfs = []
    for file in csv_files:
        try:
            dask_df = dd.read_csv(file)
            dask_dfs.append(dask_df)
        except Exception as e:
            print(f"Error reading file {file}: {e}")

    if not dask_dfs:
        raise ValueError("No valid CSV files to load.")

    combined_ddf = dd.concat(dask_dfs, axis=0)
    return combined_ddf

# Preprocess Data: Remove or transform geometry columns
def preprocess_data(dask_df):
    for col in dask_df.columns:
        print(col)
        print("type", dask_df[col].dtype )
        if col == '.geo':
            print("check2")
            # Check if the column contains JSON-like strings (e.g., geometries)
            try:
                sample_value = dask_df[col].head(1).iloc[0]
                print(sample_value)
                if isinstance(sample_value, str) and sample_value.startswith("{") and "coordinates" in sample_value:
                    print(f"Transforming geometry column: {col}")


                    # Extract the coordinates from the specified column
                    dask_df[col] = dask_df[col].apply(
                        lambda x: json.loads(x)["coordinates"] if isinstance(x, str) and "coordinates" in json.loads(x) else None,
                        meta=('x', 'object')  # 'object' because the value is a list
                    )

                    print(dask_df[col].head(1).iloc[0])
                    # Split into two numeric columns: longitude and latitude
                    dask_df[f"{col}_lon"] = dask_df[col].apply(
                        lambda x: x[0] if x else None, 
                        meta=('x', 'float')  # 'float' for numeric values
                    )
                    dask_df[f"{col}_lat"] = dask_df[col].apply(
                        lambda x: x[1] if x else None, 
                        meta=('x', 'float')  # 'float' for numeric values
                    )
                    
                    # Drop the original column to avoid redundancy
                    dask_df = dask_df.drop(columns=[col])

            except Exception as e:
                print(f"Skipping column {col}: {e}")

        # Drop remaining non-numeric columns
        elif not pd.api.types.is_numeric_dtype(dask_df[col].dtype):
            print(f"Dropping non-numeric column: {col}")
            dask_df = dask_df.drop(col, axis=1)
    return dask_df


# Convert to Xarray Dataset
def dask_to_xarray(dask_df):
    pandas_df = dask_df.compute()
    xarray_ds = xr.Dataset.from_dataframe(pandas_df)
    return xarray_ds

# Train Gradient Boosting Model
def train_model(x, y):
    X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
    model = GradientBoostingRegressor()
    model.fit(X_train, y_train)

    # Evaluate model
    y_pred = model.predict(X_test)
    rmse = mean_squared_error(y_test, y_pred, squared=False)
    print(f"RMSE: {rmse}")

    return model

# Visualize results
def visualize_results(x, y, y_pred):
    plt.figure(figsize=(10, 6))
    plt.scatter(y, y_pred, alpha=0.5, label="Predicted vs Actual")
    plt.xlabel("Actual AGB")
    plt.ylabel("Predicted AGB")
    plt.title("AGB Prediction Results")
    plt.legend()
    plt.show()

# Main execution
# def main():
print("Downloading CSV files from GCS...")
csv_files = download_csv_files()

if not csv_files:
    raise ValueError("No valid CSV files found in the specified GCS folder.")

print("Finish")
```

```python
print("Loading data with Dask...")
dask_df = load_data(csv_files)

print("Preprocessing data...")
dask_df = preprocess_data(dask_df)

print("Converting to Xarray Dataset...")
xarray_ds = dask_to_xarray(dask_df)
print("finish")
```

```python
print("Preparing data for training...")
# Extract features and target (replace 'feature_columns' and 'target_column' with actual names)
feature_columns = [col for col in xarray_ds.data_vars if col != 'agbd']
target_column = 'agbd'

X = xarray_ds[feature_columns].to_array().values.T
y = xarray_ds[target_column].values

print("finish")

```

```python
from sklearn.impute import SimpleImputer

# Handle missing values
def preprocess_missing_values(X, y):
    # Impute missing values in features
    imputer = SimpleImputer(strategy="mean")  # Replace 'mean' with 'median' or a constant if needed
    X = imputer.fit_transform(X)
    # Drop rows where target `y` is NaN
    valid_idx = ~pd.isnull(y)
    X = X[valid_idx]
    y = y[valid_idx]
    return X, y

X, y = preprocess_missing_values(X,y)

print("finish")
```

```python
print("Training Gradient Boosting Model...")
model = train_model(X, y)

print("Saving model...")
joblib.dump(model, "agb_model.pkl")

print("Visualizing results...")
y_pred = model.predict(X)
visualize_results(X, y, y_pred)
```
